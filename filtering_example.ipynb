```python
import os
import sys
from pathlib import Path
sys.path.append('./')
```

```python
import numpy as np
```

```python
import pytomofilt.filter_models as fm
import pytomofilt.model as mod
```

```python
# Import tomography model + filter files
tomo_model_spec = fm.tomographic_model_from_path(Path('./files/SP12RTS'))
ref_model = mod.RTS_Model.from_file(tomo_model_spec.coef_file)
ref_model.filter_from_file(tomo_model_spec.evec_file, tomo_model_spec.weights_file,
                            0.2,verbose=True)
```

```python
# Import synthetic (geodynamic) model (Need to improve SH coef fit speed)
geodynamic_model = mod.RTS_Model.from_directory(Path('./files/examplemodel'),12)
```

```python
# Filter synthetic (geodynamic) model
# filtered_model = ref_model.filter(geodynamic_model) 

assert isinstance(geodynamic_model, mod.RTS_Model), "Input model must be an instance of RTS_Model to be filtered" 
x_s = geodynamic_model.as_vector()
x = np.concatenate([x_s, x_s])
x = ref_model.filter_obj.apply_filter(x)
filtered_model = mod.RTS_Model(ref_model.lmax, ref_model.rmin, ref_model.rmax, ref_model.knots_r)
filtered_model.from_vector(x)
```
